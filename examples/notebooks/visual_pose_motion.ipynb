# Visual pose motion

Move to a visual pose using the visual pose library. This mirrors
`visual_pose_motion.py`.

Ensure the robot is clear before running motion commands.


```python
from r2_labs import client, rpc_api

host = "localhost"
server_address = f"tcp://{host}:{rpc_api.DEFAULT_PORT}"
query_address = f"tcp://{host}:{rpc_api.DEFAULT_QUERY_PORT}"

robot = client.Robot(server_address, query_server_address=query_address)
```

```python
visual_pose_name = "example_pose"
period_seconds = 5.0
```

```python
robot.exec_mode.set_execution_mode(rpc_api.ExecutionMode.READY)
motion_future = robot.arm.visual_pose_motion(
    visual_pose_name=visual_pose_name,
    period_seconds=period_seconds,
)
motion_future.result()
```
