# View trajectories

List trajectories and print their contents. This mirrors `view_trajectory.py`.


```python
import numpy as np

from r2_labs import client, rpc_api

host = "localhost"
server_address = f"tcp://{host}:{rpc_api.DEFAULT_PORT}"
query_address = f"tcp://{host}:{rpc_api.DEFAULT_QUERY_PORT}"
training_address=f"tcp://{host}:{rpc_api.DEFAULT_MODEL_TRAINER_PORT}"

robot = client.Robot(
    server_address,
    query_server_address=query_address,
    training_server_address=training_address,
)

np.set_printoptions(precision=4, floatmode="fixed", suppress=True, linewidth=100)
```

```python
trajectory_name = ""
```

```python
trajectories = robot.trajectory_library.list_entries()
found = False

for traj in trajectories.trajectories:
  if not trajectory_name or trajectory_name == traj.name:
    found = True
    print(f"
NAME: {traj.name}")
    print(f"TYPE: {traj.trajectory_type}")
    print(f"LENGTH (seconds): {traj.period_seconds}")

    if traj.description:
      print(f"DESCRIPTION: {traj.description}")

    print("DATA:")
    for row in traj.trajectory_data:
      print(row)

if not found and trajectory_name:
  print(f"No such trajectory: {trajectory_name}")
```
