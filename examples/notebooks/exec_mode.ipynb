# Execution mode

Query or set the robot execution mode. This mirrors `exec_mode.py`.

Select the R2 SDK kernel before running.


```python
from r2_labs import client, rpc_api

host = "localhost"
server_address = f"tcp://{host}:{rpc_api.DEFAULT_PORT}"
query_address = f"tcp://{host}:{rpc_api.DEFAULT_QUERY_PORT}"
training_address=f"tcp://{host}:{rpc_api.DEFAULT_MODEL_TRAINER_PORT}"

robot = client.Robot(
    server_address,
    query_server_address=query_address,
    training_server_address=training_address,
)
```

```python
current = robot.exec_mode.get_execution_mode()
print(f"current mode: {current.current_mode}")
```

```python
target_mode = rpc_api.ExecutionMode.READY
response = robot.exec_mode.set_execution_mode(new_mode=target_mode)
print(f"new mode: {response.current_mode}")
```

Optional: stop the robot when you are done.


```python
robot.exec_mode.set_execution_mode(new_mode=rpc_api.ExecutionMode.STOP)
```
